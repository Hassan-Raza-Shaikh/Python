```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
#from sklearn.datasets.samples_generator import make_circles
#X, y = make_circles(100, factor=.1, noise=.1)
#plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
from sklearn.datasets import make_circles
X, y = make_circles(100, factor=.1, noise=.2)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
classifier = SVC(kernel="linear",C=1.0, random_state=1)
classifier.fit(X_train, y_train.ravel())
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap

# Correct way to create the colormap
zero_one_colourmap = ListedColormap(['blue', 'red'])

def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    # Create the contour plot
    Z = clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape)
    plt.contourf(X1, X2, Z, alpha=0.75, cmap=zero_one_colourmap)
    
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    
    # Plot the scatter points
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                   color=zero_one_colourmap(i), label=j, edgecolor='black', s=80)
    
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
```

```python
plot_decision_boundary(X, y, classifier)
```

# Radial Basis Function

```python
rbf_classifier = SVC(kernel="rbf")
rbf_classifier.fit(X_train, y_train)
y_pred = rbf_classifier.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
plot_decision_boundary(X, y, rbf_classifier)
```

```python
rbf_classifier1 = SVC(kernel="rbf", random_state=1, gamma=0.10, C=10.0)
rbf_classifier1.fit(X_train, y_train)
y_pred = rbf_classifier1.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
plot_decision_boundary(X, y, rbf_classifier1)
```

```python
rbf_classifier1 = SVC(kernel="rbf", random_state=1, gamma=0.2, C=1.0)
rbf_classifier1.fit(X_train, y_train)
y_pred = rbf_classifier1.predict(X_test)
accuracy_score(y_test, y_pred)
plot_decision_boundary(X, y, rbf_classifier1)
```

```python
rbf_classifier2 = SVC(kernel="rbf", random_state=1, gamma=100.0, C=1.0)
rbf_classifier2.fit(X_train, y_train)
y_pred = rbf_classifier2.predict(X_test)
accuracy_score(y_test, y_pred)
plot_decision_boundary(X, y, rbf_classifier2)
```
