```python
class dog:
    species=None  # attribute without value
    def __init__(self, name, age):  # constructor
        self.name = name  # attribute
        self.age = age  # attribute
        return

o=dog("tommy",5)
print(o.name)
print(o.age)
```

```python
# general class for every test

class man:
    def __init__(self):  # constructor
        self.name=None  # public attribute without value
        self._age=None  # protected attribute without value
        self.__idname=None  # private attribute without value
    def display(self):  # method
        print(self.__idname)
        return
    def show(self):  # method
        print(self.idname)
        return

m=man()
```

```python
# public attribute can be accessed outside the class
print(m.name)
```

```python
# protected attribute can be accessed outside the class but not recommended
print(m._age)
```

```python
# private attribute cannot be accessed outside the class
print(m.idname)
```

```python
m.display()
m.show()
```

```python
# if i assign outside of class then it will create a new attribute/instance variable
m.age=5
print(m.age)
```

```python
# getters and setters for private attributes

class employee:
    def __init__(self):
        self.__salary=0  # private attribute without value
        return
    def set_salary(self,salary):  # setter method
        self.__salary=salary
        return
    def get_salary(self):  # getter method
        return self.__salary

e=employee()
e.set_salary(50000)  # setting private attribute
print(e.get_salary())  # getting private attribute
```

```python
# inheritance

# types of inheritance
# single inheritance
# multilevel inheritance
# multiple inheritance
# hierarchical inheritance
# hybrid inheritance
```

```python
# single inheritance

class vehicle:
    def vehicle_info(self):
        print("This is a vehicle")
        return
class car(vehicle):  # single inheritance
    def car_info(self):
        print("This is a car")
        return
c=car()
c.vehicle_info()
c.car_info()
```

```python
# multiple inheritance

class base1:
    def func1(self):
        print("This is function 1 from base 1")
        return
class base2:
    def func2(self):
        print("This is function 2 from base 2")
        return
class derived(base1,base2):  # multiple inheritance
    def func3(self):
        print("This is function 3 from derived class")
        return
d=derived()
d.func1()
d.func2()
d.func3()
```

```python
# multilevel inheritance

class grandparent:
    def func1(self):
        print("This is function 1 from grandparent class")
        return
class parent(grandparent):  # multilevel inheritance
    def func2(self):
        print("This is function 2 from parent class")
        return
class child(parent):  # multilevel inheritance
    def func3(self):
        print("This is function 3 from child class")
        return
ch=child()
ch.func1()
ch.func2()
ch.func3()
```

```python
# hierarchical inheritance

class animal:
    def func1(self):
        print("This is function 1 from animal class")
        return
class dog(animal):  # hierarchical inheritance
    def func2(self):
        print("This is function 2 from dog class")
        return
class cat(animal):  # hierarchical inheritance
    def func3(self):
        print("This is function 3 from cat class")
        return
d=dog()
d.func1()
d.func2()
c=cat()
c.func1()
c.func3()
```

```python
# hybrid inheritance

class base:
    def func1(self):
        print("This is function 1 from base class")
        return
class derived1(base):  # single inheritance
    def func2(self):
        print("This is function 2 from derived1 class")
        return
class derived2(base):  # single inheritance
    def func3(self):
        print("This is function 3 from derived2 class")
        return
class derived3(derived1,derived2):  # multiple inheritance
    def func4(self):
        print("This is function 4 from derived3 class")
        return
hd=derived3()
hd.func1()
hd.func2()
hd.func3()
hd.func4()
```

```python
# super() function

class parent:
    def func1(self):
        print("This is function 1 from parent class")
        return
class child(parent):
    def func2(self):
        super().func1()  # calling parent class method
        print("This is function 2 from child class")
        return
ch=child()
ch.func2()

# calling method of parent class
# it works like self but it is used to call parent class methods
```
