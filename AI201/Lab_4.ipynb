```python
# functions in python
```

```python
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))
```

```python
# positional arguments

def add(a, b):
    return a + b
print(add(3, 5))  # Output: 8
```

```python
# keyword arguments/ default arguments

def introduce(name='hassan', age=10):
    return f"My name is {name} and I am {age} years old."
print(introduce('bob', 25))
print(introduce())
print(introduce(age=30))    # uses default name, age=30
print(introduce(name='alice'))  # uses default age, name='alice'
```

```python
# mutable default arguments

def f(ml=[]):
    ml.append(1)
    return ml
print(f())  
print(f())
print(f())  
```

```python
# tuple unpacking

def calculate(*args):
    return sum(args)
print(calculate(1, 2, 3, 4))  # Output: 10
print(calculate(5,10,15))  # Output: 30
```

```python
def avg(*args):
    total = 0
    for i in args:
        total+=i
    return total/len(args)
print(avg(1,2,3,4,5))

l=[1,2,3,4,5]
print(avg(tuple(l)))
```

```python
def avg(*args):
    # If a single list/tuple is passed, unpack it
    if len(args) == 1 and isinstance(args[0], (list, tuple)):
        args = args[0]
    total = sum(args)
    return total / len(args)

# Examples
print(avg(1, 2, 3, 4, 5))   # 3.0
l = [1, 2, 3, 4, 5]
print(avg(l))               # 3.0
print(avg(tuple(l)))        # 3.0
```

```python
def d(**kwargs):
    print(kwargs)
    print(type(kwargs))
    for key, value in kwargs.items():
        print(f"{key}: {value}")
d(name='hassan', age=20, city='karachi')
```
