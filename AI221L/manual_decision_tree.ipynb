```
import pandas as pd
import numpy as np

# -------------------------
# FUNCTION 1: Gini Calculation
# -------------------------
def gini_index(labels):
    zeros = 0
    ones = 0

    for val in labels:
        if val == 0:
            zeros += 1
        else:
            ones += 1

    total = zeros + ones

    if total == 0:
        return 0

    p0 = zeros / total
    p1 = ones / total

    return 1 - (p0*p0 + p1*p1)


# -------------------------
# FUNCTION 2: Split Dataset
# -------------------------
def split_data(X, y, feature_index, threshold):
    left_X, right_X = [], []
    left_y, right_y = [], []

    for i in range(len(X)):
        if X[i][feature_index] <= threshold:
            left_X.append(X[i])
            left_y.append(y[i])
        else:
            right_X.append(X[i])
            right_y.append(y[i])

    return left_X, right_X, left_y, right_y


# -------------------------
# FUNCTION 3: Weighted Gini
# -------------------------
def weighted_gini(left_y, right_y):
    total = len(left_y) + len(right_y)

    gini_left = gini_index(left_y)
    gini_right = gini_index(right_y)

    weight_left = len(left_y) / total
    weight_right = len(right_y) / total

    return weight_left * gini_left + weight_right * gini_right


# -------------------------
# FUNCTION 4: Majority Class
# -------------------------
def majority_class(labels):
    zeros = 0
    ones = 0

    for val in labels:
        if val == 0:
            zeros += 1
        else:
            ones += 1

    return 0 if zeros > ones else 1


# -------------------------
# FUNCTION 5: Build 2-Level Tree
# -------------------------
def build_tree(X, y):

    best_threshold_age = 40
    left_X, right_X, left_y, right_y = split_data(X, y, 0, best_threshold_age)

    best_threshold_salary = 50000
    ll_X, lr_X, ll_y, lr_y = split_data(left_X, left_y, 1, best_threshold_salary)

    tree = {
        "feature": "Age",
        "threshold": best_threshold_age,
        "left": {
            "feature": "Salary",
            "threshold": best_threshold_salary,
            "left": majority_class(ll_y),
            "right": majority_class(lr_y)
        },
        "right": majority_class(right_y)
    }

    return tree


# -------------------------
# FUNCTION 6: Predict
# -------------------------
def predict(tree, sample):

    if sample[0] <= tree["threshold"]:

        if sample[1] <= tree["left"]["threshold"]:
            return tree["left"]["left"]
        else:
            return tree["left"]["right"]

    else:
        return tree["right"]


# -------------------------
# MAIN CODE
# -------------------------

data = pd.read_csv('Social_Network_Ads.csv')
data['Gender'] = data['Gender'].replace({'Male': 0, 'Female': 1})

X = data[['Age', 'EstimatedSalary']].values
y = data['Purchased'].values

tree = build_tree(X, y)

print("Decision Tree Structure:")
print(tree)

sample = [30, 40000]
result = predict(tree, sample)

print("\nTest Sample:", sample)
print("Prediction:", result)
```
