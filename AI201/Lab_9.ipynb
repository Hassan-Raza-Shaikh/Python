```python
# simple plot

years = [1951, 1961, 1971, 1981, 1991, 2001, 2011]
population = [36.1, 43.9, 54.8, 68.3, 84.6, 102.8, 121.0]  # in millions
plot = plt.plot(years, population) #
plt.title("Population Growth Over Years") # title of the plot
plt.xlabel("Years") # x-axis label
plt.ylabel("Population (in millions)") # y-axis label
plt.grid(True) # show grid
plt.show()
```

```python
# legend example

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
plt.plot(x, y1, label='Prime Numbers', color='blue', marker='o')
plt.plot(x, y2, label='Even Numbers', color='red', marker='x')
plt.title("Prime vs Even Numbers")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()  # show legend
plt.grid(True)
plt.show()

# labels can also be added using plt.legend(['plot1', 'plot2'],['label1', 'label2']) but for this plot needs to be stored in a variable first
```

```python
# pie chart

labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode first slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Fruit Distribution")
plt.axis('equal')  # equal aspect ratio ensures pie is drawn as a circle
plt.show()
```

```python
# subplots

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
fig, (ax1, ax2) = plt.subplots(2)  # 2 rows, 1 column
ax1.plot(x, y1, color='blue', marker='o')
ax1.set_title("Prime Numbers")
ax1.set_xlabel("Index")
ax1.set_ylabel("Value")
ax2.plot(x, y2, color='red', marker='x')
ax2.set_title("Even Numbers")
ax2.set_xlabel("Index")
ax2.set_ylabel("Value")
plt.tight_layout()  # adjust subplots to fit into figure area.
plt.show()

# subplot returns a figure and axes object which can be used to customize individual plots
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import seaborn as sns

# --------------------------------------------------
# 1. Data Preparation (exactly as in the challenge)
# --------------------------------------------------
np.random.seed(42)

classes = ['Class 9A', 'Class 9B', 'Class 9C', 'Class 10A', 'Class 10B', 'Class 10C']

data = {
    'Class': classes,
    'Math':    [78, 85, 69, 92, 88, 81],
    'Science': [82, 88, 74, 90, 85, 79],
    'English': [75, 80, 72, 85, 89, 83],
    'History': [88, 76, 81, 79, 84, 87],
    'Students': [40, 35, 30, 45, 38, 42],
    'Avg_Attendance_%': [94, 91, 88, 96, 93, 90],
    'Failed_Students_Math': [6, 4, 12, 2, 5, 8],
}

df = pd.DataFrame(data)

# Individual student scores
students = []
for i, row in df.iterrows():
    n = row['Students']
    math_scores = np.clip(np.random.normal(row['Math'], 12), 0, 100).astype(int)
    science_scores = np.clip(0.85 * math_scores + np.random.normal(0, 8, n), 0, 100).astype(int)
    students.extend(zip([row['Class']]*n, math_scores, science_scores))

student_df = pd.DataFrame(students, columns=['Class', 'Math_Score', 'Science_Score'])

# Extracurricular (total and Class 10 only)
total_activities = {'Sports': 135, 'Music/Arts': 82, 'Debate': 48, 'Robotics': 35, 'None': 28}
class10_activities = {'Sports': 68, 'Music/Arts': 38, 'Debate': 25, 'Robotics': 22, 'None': 10}  # made realistic

# --------------------------------------------------
# 2. Main Dashboard  23 subplots
# --------------------------------------------------
plt.style.use('seaborn-v0_8-whitegrid')
fig, axs = plt.subplots(2, 3, figsize=(18, 11), constrained_layout=False)
fig.suptitle('School Performance Dashboard  Academic Year 20242025',
              fontsize=20, fontweight='bold', y=0.98)

# Colors
colors_subjects = sns.color_palette("husl", 4)
colors_attendance = ['#2E8B57', '#CD5C5C']
colors_activities_outer = sns.color_palette("Set2", 5)
colors_activities_inner = sns.color_palette("Pastel1", 5)

# 1. Grouped Bar Chart  All subjects
x = np.arange(len(classes))
width = 0.2
axs[0,0].bar(x - 1.5*width, df['Math'],    width, label='Math',    color=colors_subjects[0])
axs[0,0].bar(x - 0.5*width, df['Science'], width, label='Science', color=colors_subjects[1])
axs[0,0].bar(x + 0.5*width, df['English'], width, label='English', color=colors_subjects[2])
axs[0,0].bar(x + 1.5*width, df['History'], width, label='History', color=colors_subjects[3])
axs[0,0].set_title('Average Scores by Subject and Class', fontsize=14, fontweight='bold')
axs[0,0].set_xticks(x)
axs[0,0].set_xticklabels(classes, rotation=30)
axs[0,0].set_ylabel('Average Score')
axs[0,0].legend()
axs[0,0].grid(axis='y', alpha=0.3)

# 2. Horizontal 100% Stacked Bar  Attendance
axs[0,1].barh(classes, df['Avg_Attendance_%'], color=colors_attendance[0], label='Present')
axs[0,1].barh(classes, 100 - df['Avg_Attendance_%'], left=df['Avg_Attendance_%'],
              color=colors_attendance[1], label='Absent')
axs[0,1].set_title('Attendance Rate per Class (100% stacked)', fontsize=14, fontweight='bold')
axs[0,1].set_xlabel('Percentage (%)')
axs[0,1].legend()
axs[0,1].invert_yaxis()

# 3. Bar with error bars + annotation (Math failures)
failures = df['Failed_Students_Math']
errors = np.sqrt(failures)  # rough standard error
bars = axs[0,2].bar(classes, failures, yerr=errors, capsize=5,
                   color=sns.color_palette("Reds", 6), edgecolor='black')
axs[0,2].set_title('Students Who Failed Math', fontsize=14, fontweight='bold')
axs[0,2].set_ylabel('Number of Students')
# Annotate worst class
worst_idx = failures.idxmax()
axs[0,2].annotate('Worst performing!', xy=(worst_idx, failures[worst_idx]),
                  xytext=(worst_idx, failures[worst_idx]+4),
                  arrowprops=dict(arrowstyle="->", color='black'),
                  ha='center', fontsize=12, color='red', fontweight='bold')

axs[0,2].tick_params(axis='x', rotation=30)

# 4. Scatter Math vs Science + trend line
scatter = axs[1,0].scatter(student_df['Math_Score'], student_df['Science_Score'],
                     c=pd.Categorical(student_df['Class']).codes, cmap='tab10', alpha=0.7)
axs[1,0].set_title('Math vs Science Individual Scores', fontsize=14, fontweight='bold')
axs[1,0].set_xlabel('Math Score')
axs[1,0].set_ylabel('Science Score')
# Trend line
z = np.polyfit(student_df['Math_Score'], student_df['Science_Score'], 1)
p = np.poly1d(z)
axs[1,0].plot(student_df['Math_Score'], p(student_df['Math_Score']), "k--", lw=2)
# Correlation
corr = student_df[['Math_Score', 'Science_Score']].corr().iloc[0,1].round(3)
axs[1,0].text(0.05, 0.95, f'Pearson r = {corr}', transform=axs[1,0].transAxes,
             bbox=dict(boxstyle="round", facecolor="wheat"), fontsize=12)

# 5. Violin plot of Math scores per class
parts = axs[1,1].violinplot([student_df[student_df['Class']==c]['Math_Score'].values for c in classes],
                           positions=range(len(classes)), widths=0.8)
for pc in parts['bodies']:
    pc.set_facecolor('#4C72B0')
    pc.set_alpha(0.8)
axs[1,1].set_title('Distribution of Math Scores by Class', fontsize=14, fontweight='bold')
axs[1,1].set_xticks(range(len(classes)))
axs[1,1].set_xticklabels(classes, rotation=30)
axs[1,1].set_ylabel('Math Score')

# 6. Nested Donut Chart
size = 0.3
outer_vals = list(total_activities.values())
inner_vals = list(class10_activities.values())

axs[1,2].pie(outer_vals, radius=1, colors=colors_activities_outer,
            labels=total_activities.keys(), wedgeprops=dict(width=size, edgecolor='w'),
            startangle=90, labeldistance=0.75, textprops={'fontsize': 10})

axs[1,2].pie(inner_vals, radius=1-size, colors=colors_activities_inner,
            labels=class10_activities.keys(), wedgeprops=dict(width=size, edgecolor='w'),
            startangle=90, labeldistance=0.7, textprops={'fontsize': 9})

axs[1,2].set_title('Extracurricular Participation\n(Outer: All | Inner: Class 10 only)',
                   fontsize=14, fontweight='bold')

# Final adjustments
fig.tight_layout()
fig.subplots_adjust(top=0.92)
plt.savefig('school_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

# --------------------------------------------------
# 7. Bonus: Correlation Heatmap (standalone)
# --------------------------------------------------
corr_df = df[['Math', 'Science', 'English', 'History', 'Avg_Attendance_%']].corr()

plt.figure(figsize=(8, 6))
mask = np.triu(np.ones_like(corr_df, dtype=bool))
plt.imshow(corr_df, cmap='coolwarm', vmin=-1, vmax=1, mask=mask)
plt.colorbar(label='Correlation Coefficient')

# Annotate values
for i in range(len(corr_df)):
    for j in range(len(corr_df)):
        if not mask[i,j]:
            plt.text(j, i, f'{corr_df.iloc[i,j]:.2f}',
                     ha='center', va='center', color='white' if abs(corr_df.iloc[i,j])>0.6 else 'black',
                     fontsize=12, fontweight='bold')

plt.xticks(range(len(corr_df.columns)), corr_df.columns, rotation=45)
plt.yticks(range(len(corr_df.columns)), corr_df.columns)
plt.title('Subject & Attendance Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
