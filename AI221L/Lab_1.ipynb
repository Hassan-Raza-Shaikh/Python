```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Hours' : [1,2,3,4,5],
    'Marks' : [35,40,50,55,65]
}

df=pd.DataFrame(data)
print(df)
```

```python
x = df['Hours'].values
y = df['Marks'].values
print("X:",x)
print("Y:",y)
```

```python
plt.scatter(x,y)
plt.xlabel("Hours studied")
plt.ylabel("Marks obtained")
plt.title("Hours vs Marks")
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print("Mean of X:", x_mean)
print("Mean of Y:", y_mean)
```

```python
numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x - x_mean)**2)
a= numerator / denominator
print("Slope (a):",a)
```

```python
b = y_mean - a * x_mean
print("Y-Intercept (b):",b)
```

```python
print(f"Linear Regression Equation: Y = {a}X + {b}")
```

```python
y_pred = a*x + b
print("Predicted Y values:", y_pred)
```

```python
z = np.array([4,5,6,7])
y_pred_new = a*z + b
print("Predicted Y values for new X values:", y_pred_new)
```

```python
plt.scatter(x,y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Regression Line')
plt.xlabel("Hours studied")
plt.ylabel("Marks obtained")
plt.legend()
plt.show()
```

```python
errors = y-y_pred
print("Residuals:", errors)
```

```python
mse = np.mean(errors**2)
print("Mean Squared Error (MSE):", mse)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x_reshaped = x.reshape(-1, 1)
model.fit(x_reshaped, y)
y_sklearn_pred = model.predict(x_reshaped)
mse_sklearn = mean_squared_error(y, y_sklearn_pred)
print("Mean Squared Error (sklearn):", mse_sklearn)
print("Predicted Y values using sklearn:", y_sklearn_pred)
```
