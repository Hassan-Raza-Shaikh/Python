```python
# escape characters
print('I\'m a student')
print("He said \"Hello\" to me")
print("This is line 1\nThis is line 2")
print("This is line 1\tThis is line 2")
print("This is a backslash: \\")
print('a \n b \
c')
print('\a')
print('\u1111') # unicode character
```
